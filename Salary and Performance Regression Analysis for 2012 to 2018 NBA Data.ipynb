# Salary and Performance Regression Analysis for 2012 to 2018 NBA Data

## Import neccesary modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
```

## Download NBA Salary data from 2012 to 2018

```python
NBA=pd.read_excel("12to18 NBA pay and performance.xlsx")
```

## Desriptive Statistics for NBA Salary df

```python
NBA.describe()
```

## Calculate total NBA salary for each season

```python
Sumsal = NBA.groupby(['season'])['salaries'].sum().reset_index().rename(columns={'salaries':'allsal'})
Sumsal
```

```python
NBA = pd.merge(NBA, Sumsal, on=['season'], how='left')
display(NBA)
```

## Calculate relative salary for all teams and seaons 

```python
NBA['relsal']= NBA['salaries']/NBA['allsal']
```

## Plot relative salary against winning percentage

```python
sns.regplot(x="relsal", y="wpc", data = NBA, ci=False)
```

```python
with sns.axes_style('white'):
    sns.jointplot(x="relsal", y="wpc", data=NBA, kind='hex')
```

### Run Ordinary Least Squares Linear Regressions for WPC vs Relsal

```python
wpcsal1_lm = smf.ols(formula = 'wpc ~ relsal', data=NBA).fit()
print(wpcsal1_lm.summary())
```

```python
NBA.sort_values(by=['Team','season'], ascending=True)
```

## Insert column for previous season's winning percentage

```python
NBA['wpc_lag'] = NBA.groupby('Team')['wpc'].shift(1)
NBA
```

```python
pd.set_option('display.max_rows', 250)
NBA
```

### Run Ordinary Least Squares Linear Regressions for WPC vs Relsal + WPC_lag

```python
wpcsal2_lm = smf.ols(formula = 'wpc ~ wpc_lag + relsal', data=NBA).fit()
print(wpcsal2_lm.summary())
```

### Run Ordinary Least Squares Linear Regressions for WPC vs Relsal + WPC_lag + Team column

```python
wpcsal3_lm = smf.ols(formula = 'wpc ~ wpc_lag + relsal +C(Team)', data=NBA).fit()
print(wpcsal3_lm.summary())
```
